```python
import yfinance as yf
import datetime as dt
stock = "POWERGRID.NS"
start = dt.datetime(2000,1,1) 
end = dt.datetime(2025,2,10)  
df = yf.download(stock,start,end)
```

```python
data01 = pd.read_csv("powergrid.csv")
```

```python
data01.head()
```

```python

# สร้าง Candlestick Chart
fig = go.Figure(data=[go.Candlestick(
    x=data01['Date'],               # วันที่ (Date)
    open=data01['Open'],            # ราคาต้น (Open)
    high=data01['High'],            # ราคาสูงสุด (High)
    low=data01['Low'],              # ราคาต่ำสุด (Low)
    close=data01['Close']           # ราคาปิด (Close)
)])
# แสดงกราฟ
fig.show()
```

```python
df = df.drop(["Date"], axis=1)
```

plot กราฟราคาปิด✅

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label = f'{stock} Closing Price', linewidth = 1)
plt.title(f'{stock} Closing prices over time')
plt.legend()
plt.show()
```

plot กราฟราคาเปิด✅

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Open'], label = f'{stock} Opening Price', linewidth = 1)
plt.title(f'{stock} Opening prices over time')
plt.legend()
plt.show()
```

plot กราฟราคาall time hight✅

```python
plt.figure(figsize=(12, 6))
plt.plot(df['High'], label = f'{stock} High Price', linewidth = 1)
plt.title(f'{stock} High prices over time')
plt.legend()
plt.show()
```

plot กราฟวอลุ่มของหุ้น✅

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Volume'], label = f'{stock} Volume', linewidth = 2)
plt.title(f'{stock} Volume over time')
plt.legend()
plt.show()
```
