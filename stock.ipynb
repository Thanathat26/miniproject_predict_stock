```python
import pandas as pd
ma5 = df.Close.rolling(5).mean()
ma10 = df.Close.rolling(10).mean()
ma50 = df.Close.rolling(50).mean()
ma100 = df.Close.rolling(100).mean()
ma200 = df.Close.rolling(200).mean()
ma5,ma10,ma50,ma100,ma200
```

สร้างเส้น sma✅

```python
plt.figure(figsize=(12, 6))
plt.plot(df.Close, label = f'{stock} Close Price', linewidth = 1)
plt.plot(ma5, label = f'{stock} Moving Average 5 Price', linewidth = 1)
plt.plot(ma10, label = f'{stock} Moving Average 10 Price', linewidth = 1)
plt.plot(ma50, label = f'{stock} Moving Average 50 Price', linewidth = 1)
plt.plot(ma100, label = f'{stock} Moving Average 100 Price', linewidth = 1)
plt.plot(ma200, label = f'{stock} Moving Average 200 Price', linewidth = 1)
plt.legend()
plt.show()
```

สร้างเส้น ema✅

```python
ema5 = df.Close.ewm(span=5, adjust = False).mean()
ema10 = df.Close.ewm(span=10, adjust = False).mean()
ema50 = df.Close.ewm(span=50, adjust = False).mean()
ema100 = df.Close.ewm(span=100, adjust = False).mean()
ema200 = df.Close.ewm(span=200, adjust = False).mean()
ema5,ema10,ema50,ema100,ema200
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.Close, label = f'{stock} Close Price', linewidth = 1)
plt.plot(ema5, label = f'{stock} Moving Average 5 Price', linewidth = 1)
plt.plot(ema10, label = f'{stock} Moving Average 10 Price', linewidth = 1)
plt.plot(ema50, label = f'{stock} Moving Average 50 Price', linewidth = 1)
plt.plot(ema100, label = f'{stock} Moving Average 100 Price', linewidth = 1)
plt.plot(ema200, label = f'{stock} Moving Average 200 Price', linewidth = 1)
plt.legend()
plt.show()
```
